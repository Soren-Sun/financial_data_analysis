```python
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif'] = 'KaiTi'
plt.rcParams['axes.unicode_minus'] = False
import numpy as np
import pandas as pd
file_path=r"E:\数据表格\数据表格\《Python金融实战案例精粹（第2版）》的数据Excel文件\第4章\4.2 案例数据\货币供应量及人民币贷款余额数据.xlsx"
df=pd.read_excel(file_path,index_col=('日期'))
df.index=pd.to_datetime(df.index)
df.describe()
```
