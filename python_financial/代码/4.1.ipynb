```python
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif'] = 'KaiTi'
plt.rcParams['axes.unicode_minus'] = False
import numpy as np
import pandas as pd
file_path=r"E:\数据表格\数据表格\《Python金融实战案例精粹（第2版）》的数据Excel文件\第4章\4.1 案例数据\2021年5月18日主要期限的国债到期收益率.xlsx"
df=pd.read_excel(file_path)
df
```

```python
arr1=np.array(df['期限（年）'])
arr2=np.array(df['收益率'])
```

```python
result_1=np.polyfit(arr1,arr2,deg=1)
fun_1=np.poly1d(result_1)
rate_list1=fun_1(arr1)
rate_list1
```

```python
result_2=np.polyfit(arr1,arr2,deg=2)
fun_2=np.poly1d(result_2)
rate_list2=fun_2(arr1)
rate_list2
```

```python
result_3=np.polyfit(arr1,arr2,deg=3)
fun_3=np.poly1d(result_3)
rate_list3=fun_3(arr1)
rate_list3
```

```python
fig=plt.figure(figsize=(16,9))
plt.scatter(df['期限（年）'],df['收益率'],s=16,c='r',label='收益率',marker='o',lw=(2.5))
plt.plot(arr1,rate_list1,'b',label='一阶拟合',linestyle='--')
plt.plot(arr1,rate_list2,'g',label='二阶拟合',linestyle='-')
plt.plot(arr1,rate_list3,'k',label='三阶拟合',linestyle='-.')
plt.grid()
plt.legend(fontsize=16)
plt.xlabel('期限',fontsize=16)
plt.ylabel('收益率',fontsize=16)
plt.title('拟合',fontsize=20)
```
