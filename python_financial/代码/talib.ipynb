```python
# ma5 = talib.MA(df['close'], timeperiod=5)
# ma10 = talib.MA(df['close'], timeperiod=10)
# ma20 = talib.MA(df['close'], timeperiod=20)
# fig=plt.figure(figsize=(16,9))
# ax=fig.add_subplot(111)
# ax.plot(ma5,label='MA5')
# ax.plot(ma10,label='M10')
# ax.plot(ma20,label='MA20')
# mpf.plot(df)
import talib
import matplotlib.pyplot as plt
import mplfinance as mpf

# Calculate moving averages
ma5 = talib.MA(df['close'], timeperiod=5)
ma10 = talib.MA(df['close'], timeperiod=10)
ma20 = talib.MA(df['close'], timeperiod=20)

# Create a plot using mplfinance
apds = [
    mpf.make_addplot(ma5, color='blue', label='MA5'),
    mpf.make_addplot(ma10, color='orange', label='MA10'),
    mpf.make_addplot(ma20, color='green', label='MA20')
]

# Plot the candlestick chart with moving averages
mpf.plot(df, type='candle', addplot=apds, figsize=(16, 9), style='charles', volume=False)
```
